```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_school= len(school_data_complete["school_name"].unique())
total_students = school_data_complete["student_name"].count()
total_budget = school_data_complete["budget"].unique().sum()
average_math = school_data_complete["math_score"].mean()
average_read = school_data_complete["reading_score"].mean()
overall_avg = (average_math + average_read)/2
math_stats = school_data_complete.loc[school_data_complete["math_score"] >= 70]
math_percent = len(math_stats) /school_data_complete["Student ID"].count()
read_stat = school_data_complete.loc[school_data_complete["reading_score"]>= 70]
read_percent = len(read_stat)/school_data_complete["Student ID"].count()
summary_table = pd.DataFrame({"Total\nSchool": [total_school],
                              "Total\nStudents": [total_students],
                              "Total Budget": [total_budget],
                              "Average Math\nScore": [average_math],
                              "Average Reading\nScore": [average_read],
                              "%Passing\nMath": [math_percent],
                              "%Passing\nReading": [read_percent],
                              "%Overall Passing\nRate": [overall_avg]})

diff_summary = summary_table.reindex_axis(["Total\nSchool",
                             "Total\nStudents",
                             "Total Budget",
                             "Average Math\nScore",
                             "Average Reading\nScore",
                             "%Passing\nMath",
                             "%Passing\nReading",
                             "%Overall Passing\nRate"], axis=1)
diff_summary.style.format({"Total\nStudents": "{:,}",
                           "Total Budget": "${:,.2f}",
                          "Average Math\nScore": "{:.1f}",
                           "Average Reading\nScore": "{:.1f}",
                          "%Passing\nMath": "{:%}",
                          "%Passing\nReading": "{:%}",
                          "%Overall Passing\nRate": "{:.2f}"})
```

```python
school_def = school_data_complete.groupby(["school_name"])
school_type = school_def["type"].first()
studs_school = school_def["student_name"].count()
tot_budget = school_def["budget"].first()
avg_stu_bud = school_def["budget"].first()/school_def["student_name"].count()
avg_math = school_def["math_score"].mean()
avg_read = school_def["reading_score"].mean()
math_stat = school_data_complete.loc[school_data_complete["math_score"] >= 70].groupby(["school_name"]).size()
math_per = math_stat /studs_school*100
read_stats = school_data_complete.loc[school_data_complete["reading_score"]>= 70].groupby(["school_name"]).size()
read_perc = read_stats/studs_school*100
over_perc = (math_per + read_perc)/2

summ_table = pd.DataFrame({"School\nType": school_type,
                              "Total\nStudents": studs_school,
                              "Total Budget": tot_budget,
                           "Per Student\nBudget": avg_stu_bud,
                              "Average Math\nScore": avg_math,
                              "Average Reading\nScore": avg_read,
                              "%Passing\nMath": math_per,
                              "%Passing\nReading": read_perc,
                              "%Overall Passing\nRate": over_perc})

diff_summ = summ_table.reindex_axis(["School\nType",
                             "Total\nStudents",
                             "Total Budget",
                            "Per Student\nBudget",
                             "Average Math\nScore",
                             "Average Reading\nScore",
                             "%Passing\nMath",
                             "%Passing\nReading",
                             "%Overall Passing\nRate"], axis=1)
```
